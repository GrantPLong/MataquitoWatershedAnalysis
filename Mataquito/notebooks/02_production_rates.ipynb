# Calculating production rates for CT samples

## Overview

**Purpose**: Calculate cosmogenic ¹⁰Be surface production rates for each sample site using the Stone (2000) latitude/altitude scaling scheme.

**Inputs**:
- Sample elevations from `MataquitoSampleData.xlsx` (via `mataquito.production_rates`)

**Outputs**:
- Spallation-only surface production rates (atoms/g/yr) for CT-1 through CT-11

**Method**: Uses the 30-40° latitude bin polynomial coefficients from Stone (2000) to convert sample-averaged elevation → atmospheric pressure → scaled production rate relative to the sea-level high-latitude (SLHL) reference of 5.1 atoms/g/yr.

```python
import numpy as np
import sys; sys.path.insert(0, '..')

from mataquito.production_rates import (
    stone2000_production_rate, elevation_to_pressure,
    SAMPLE_ELEVATIONS, STONE_COEFFICIENTS, P_SLHL,
)
```

```python
# P = N * E * p / capital lambda

atten = 160 # g/cm^2 attenuation length. 

p = 2.6 # g/cm^3 density of rock

N = np.array([5.45E+05, 8.46E+04, 1.15E+05, 2.98E+05, 2.88E+04, 3.03E+05, 2.33E+04, 3.38E+04, 7.34E+04, 3.64E+04, 3.11E+04])
Nu = np.array([6.58E+03, 1.61E+03, 2.15E+03, 5.92E+03, 6.86E+02, 5.70E+03, 1.34E+03, 1.24E+03, 1.40E+03, 1.28E+03, 1.04E+03])
E = np.array([22.9, 94.8, 29.5, 29.6, 387, 29.8, 532, 263, 105, 246, 286])
Eu = np.array([])

Ecm = (E * 100) / 1000000 # erosion rate in cm/yr

print("Production rates (atm/g/yr)")
P = (N * Ecm * p) / atten

samples = ['CT-1', 'CT-2', 'CT-3', 'CT-4', 'CT-5', 'CT-6', 'CT-7', 'CT-8', 'CT-9', 'CT-10', 'CT-11']
for sample, prod_rate in zip (samples, P):
    print(f"{sample} {prod_rate}")
```

```python
# Stone (2000) surface production rates using the mataquito package
print("Surface production rates (atoms/g/yr) - Spallation only:")
print("-" * 50)

for name, elev in SAMPLE_ELEVATIONS.items():
    P_surf = stone2000_production_rate(elev)
    print(f"{name:12s}: {P_surf:7.2f}")
```
